```python
# IMPORTANDO BIBLIOTECAS
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pdb
from yahoofinancials import YahooFinancials
import statsmodels.api as sm
```

```python
# URL DA API DO IBGE SIDRA PARA IMPORTAR TABELA
# OS DADOS DEPOIS DO .../value/... SÃO AS CONFIG DO API
url_sidra = "http://api.sidra.ibge.gov.br/values/t/1620/n1/all/v/all/p/last%2077/c11255/93404/d/v583%204"

# IMPORTANDO TABELA EM JSON E CRIANDO DATAFRAME
sidra_df = pd.DataFrame(requests.get(url=url_sidra).json())
sidra_df.columns = sidra_df.iloc[0]
sidra_df = sidra_df.iloc[1:]
sidra_df
```

```python
# CALCULANDO ARRAY COM O RETORNO DO CONSUMO UTILIZANDO NUMPY
# FORMULA DA CONTA: LOG(Ct/Ct-1)
cons_ret_array = np.log(np.array(pd.to_numeric(sidra_df.iloc[1:, 10])) / np.array(pd.to_numeric(sidra_df.iloc[:-1, 10])))
cons_ret_df = pd.DataFrame(cons_ret_array, columns=['Retorno'], index=sidra_df.iloc[1:, 5])
cons_ret_df
```

```python
# ACOES A SEREM BUSCADAS NO YAHOO FINANCE
blue_chips = ['PETR3.SA', 'GGBR3.SA', 'ABEV3.SA', 'BBAS3.SA', 'CMIG4.SA']

# QUERYING CONFIG
start_date = '2000-01-01'  # DATA INICIAL DA SERIE PARA BUSCAR NO YAHOO
end_date = '2018-12-31'  # DATA FINAL DA SERIE PARA BUSCAR NO YAHOO
numb_of_tri = ((int(end_date[:4]) - int(start_date[:4]))+1)*4 # CALCULA NUMERO DE TRIMESTRE CONFORME DATAS
period = 'daily'  # TIPO DE DADO, PODE SER 'daily', 'weekly' and 'monthly'
price_type = 'adjclose'  # PODE SER close OR adjclose (AJUSTADO PARA AMBOS DIVIDENDOS E SPLITS)

# QUARTER START AND END MONTHS
dates_dict = {
    1: ['01', '03'],
    2: ['04', '06'],
    3: ['07', '09'],
    4: ['10', '12']
}
```

```python
# BUSCANDO NO YAHOO FINANCA OS DADOS
yahoo_financials_blue_chips = YahooFinancials(blue_chips)
blue_chips_dict = yahoo_financials_blue_chips.get_historical_price_data(start_date, end_date, period)
del yahoo_financials_blue_chips
```

```python
lst_returns = []

# CALCULANDO RETORNO PARA CADA TICKER
for ticker in blue_chips:
    lst_dates = []
    lst_returns_temp = []
    df1 = pd.DataFrame(blue_chips_dict[ticker]['prices'])
    if int(df1['formatted_date'].iloc[-1][:4]) < int(df1['formatted_date'].iloc[0][:4]):
        pdb.set_trace()
        df1 = df1.iloc[::-1]

    if int(df1['formatted_date'].iloc[0][:4]) == 2000 and int(df1['formatted_date'].iloc[0][5:7]) == 1:

        lst_dates.append(ticker + '_stock_date')
        for year in range(2000, 2019):
            for sem in range(1, 5):
                for dia in range(1, 32):
                    if dia < 10:
                        data_ = str(year) + '-' + dates_dict[sem][0] + '-0' + str(dia)
                    else:
                        data_ = str(year) + '-' + dates_dict[sem][0] + '-' + str(dia)
                    try:
                        vol_ = df1[df1['formatted_date'] == data_]['volume'].values[0]
                        try:
                            vol_
                        except:
                            pass
                        else:
                            if vol_ > 0:
                                price_1 = df1[df1['formatted_date'] == data_][price_type].values[0]
                                price_1_date = df1[df1['formatted_date'] == data_]['formatted_date'].values[0]
                                del vol_
                                break
                            else:
                                pass
                    except:
                        pass

                for dia in range(31, 1, -1):
                    data_ = str(year) + '-' + dates_dict[sem][1] + '-' + str(dia)
                    try:
                        vol_ = df1[df1['formatted_date'] == data_]['volume'].values[0]
                        try:
                            vol_
                        except:
                            pass
                        else:
                            if vol_ > 0:
                                price_2 = df1[df1['formatted_date'] == data_][price_type].values[0]
                                price_2_date = df1[df1['formatted_date'] == data_]['formatted_date'].values[0]
                                del vol_
                                break
                            else:
                                pass
                    except:
                        pass

                try:
                    ln_return = np.log(price_2 / price_1)
                    lst_returns_temp.append(ln_return)
                    lst_dates.append([price_1_date, price_2_date])
                except:
                    print(
                        "----------------------------------------------------------------------------------------------------------------")
                    print('There was a problem calculating the ln return of', ticker,
                          ', problably caused by a inconsistency with the data ')
                    print(
                        "----------------------------------------------------------------------------------------------------------------")
                    pdb.set_trace()

        if len(lst_returns_temp) != numb_of_tri:
            print("The number of returns is different than", numb_of_tri, "for", ticker)
            pdb.set_trace()
        # lst_returns.append(lst_dates)
        lst_returns.append(lst_returns_temp)
        del lst_dates, lst_returns_temp

    else:
        print("---------------------------------------------------")
        print(ticker, 'does not have enought data!')
        print("---------------------------------------------------")
```

```python
# CRIANDO DATAFRAME COM RETORNOS TRIMESTRAIS DAS ACOES
stock_return_df = pd.DataFrame(lst_returns).T
stock_return_df.columns = blue_chips
stock_return_df.index = sidra_df.iloc[1:, 5]
```

```python
# UNINDO TODOS EM UM MESMO DATAFRAME

final_df = stock_return_df
final_df['Consumo'] = cons_ret_df.iloc[:, 0].tolist()
final_df['Const'] = 1
final_df = final_df.reset_index()
```

```python
#final_df = final_df.iloc[-40:]
```

```python
reg1 = sm.OLS(endog=final_df['PETR3.SA'], exog=final_df[['Const', 'Consumo']], missing='drop')
type(reg1)
```

```python
results = reg1.fit()
type(results)
results.summary()
```

```python
results_summary = results.summary()

# Note that tables is a list. The table at index 1 is the "core" table. Additionally, read_html puts dfs in a list, so we want index 0
results_as_html = results_summary.tables[1].as_html()
df_results1 = pd.read_html(results_as_html, header=0, index_col=0)[0]

results_as_html = results_summary.tables[0].as_html()
df_results2 = pd.read_html(results_as_html, header=0, index_col=0)[0]

df_results2.reset_index()

df_results1[df_results2.keys()[1]] = df_results2.keys()[2]
df_results1[df_results2.iloc[0,1]] = df_results2.iloc[0,2]
df_results1

```
